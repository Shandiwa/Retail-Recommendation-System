```python
# Retail Recommendation Engine

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Sample customer purchase data
data = {
    'Customer': ['A','A','B','B','C','C','D','D'],
    'Product': ['Laptop','Mouse','Laptop','Keyboard','Mouse','Keyboard','Laptop','Mouse'],
    'Rating': [5,4,5,3,4,5,4,5]
}

df = pd.DataFrame(data)

# Create User-Item Matrix
user_item = df.pivot_table(index='Customer',
                           columns='Product',
                           values='Rating',
                           fill_value=0)

print("User-Item Matrix")
print(user_item)

# Cosine Similarity
similarity = cosine_similarity(user_item)

similarity_df = pd.DataFrame(similarity,
                             index=user_item.index,
                             columns=user_item.index)

print("\nCustomer Similarity Matrix")
print(similarity_df)
```
